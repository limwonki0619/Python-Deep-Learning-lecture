## 2. 초음파 광물 예측하기 : 학습셋과 테스트셋 구분

```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# seed 값 설정

seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)

df = pd.read_csv('../dataset/sonar.csv', header=None)

dataset = df.values
X = dataset[:, 0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 학습셋과 테스트 셋의 구분

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=130, batch_size=5)
```

```python
# 테스트셋에 모델 적용
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```

## 4. 초음파 광물 예측하기 : K겹 교차 검증

- 머신러닝 중 직면하는 문제는 테스트셋을 전체 데이터의 30% 밖에 사용하지 못한다는 점이다.
- 이러한 단점을 보완하고자 만든 방법이 **K겹 교차 검증(k-fold cross validation)**이다.
- K겹 교차검증이란 데이터셋을 여러개로 나누어 하나씩 테스트셋으로 사용하고, 나머지를 학습셋으로 사용하는 방법이다.

```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# seed값 설정

seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)

df = pd.read_csv('../dataset/sonar.csv', header=None)

dataset = df.values    # 배열로 생성 
X = dataset[:, 0:60]   # ndarray[행, 열], 0 ~ 59열 
Y_obj = dataset[:,60]  # 60열  

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)  # 'R'과 'M'이었던 라벨을 0과 1로 변환 
```

```python
# 10개의 파일로 쪼갬

n_fold = 10
skf = StratifiedKFold(n_splits = n_fold, shuffle=True, random_state=seed)  # Stratified : 계층화하다. 층을 이루다
```

```python
# 모델의 설정, 컴파일 실행

accuracy = []  # 빈 accuracy 리스트 생성 

for train, test in skf.split(X, Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
    k_accuracy = '%.4f' % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)
```

```python
print(' \n %.f fold accuracy:' % n_fold, accuracy)
```
