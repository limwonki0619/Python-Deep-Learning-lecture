## 와인의 종류 예측하기 : 데이터의 확인과 실행

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy 
import tensorflow as tf
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.set_random_seed(seed)

# 데이터 입력
df_pre = pd.read_csv('../dataset/wine.csv', header=None)
df = df_pre.sample(frac=1)

dataset = df.values
X = dataset[:,0:12]
Y = dataset[:,12]
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(30,  input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#모델 컴파일
model.compile(loss='binary_crossentropy',
           optimizer='adam',
           metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=200)
```

```python
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```

## 와인의 종류 예측하기 : 모델 업데이트

```python
from keras.callbacks import ModelCheckpoint
import os
```

```python
# 모델 저장 폴더 설장 
MODEL_DIR = './model/'

if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
```

```python
# 모델 저장 조건 설정 
modelpath="./model/{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
```

```python
model.fit(X, Y, validation_split=0.2, epochs=200, batch_size=200, verbose=0, callbacks=[checkpointer])
```

## 와인의 종류 예측하기 : 그래프 표현

```python
# 모델 실행 및 저장 
history = model.fit(X, Y, validation_split = 0.33, epochs = 3500, batch_size = 500)
```

```python
# 테스트셋으로 실험 결과의 오차 값을 저장 
y_vloss = history.history['val_loss']
```

```python
# 학습셋으로 측정한 정확도의 값을 지정 
y_acc = history.history['acc']
```

```python
x_len = numpy.arange(len(y_acc))

plt.plot(x_len, y_vloss, 'o', c='red', markersize=3)
plt.plot(x_len, y_acc, 'o', c="blue", markersize=3)
plt.show()
```

## 와인의 종류 예측하기 : 학습의 자동 중단

```python
from keras.callbacks import EarlyStopping 
```

```python
# 자동 중단 설정 
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)
```

```python
model.fit(X, Y, validation_split=0.2, epochs=2000, batch_size=500, callbacks=[early_stopping_callback])
```

```python
print('\n Accuracy : %.4f' % (model.evaluate(X, Y)[1]))
```

## 와인 종류 예측하기 : 전체 코드

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy
import os
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.set_random_seed(seed)

df_pre = pd.read_csv('../dataset/wine.csv', header=None)
df = df_pre.sample(frac=0.15)  # 전체 샘플 중 15%만 사용 

dataset = df.values  # 배열로 변경 
X = dataset[:, 0:12]
Y = dataset[:, 12]

model = Sequential()
model.add(Dense(30,  input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
          optimizer='adam',
          metrics=['accuracy'])

# 모델 저장 폴더 만들기
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
   os.mkdir(MODEL_DIR)

modelpath="./model/{epoch:02d}-{val_loss:.4f}.hdf5"

# 모델 업데이트 및 저장
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)

# 학습 자동 중단 설정
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)

model.fit(X, Y, validation_split=0.2, epochs=3500, batch_size=500, verbose=0, callbacks=[early_stopping_callback,checkpointer])

```

```python
print("\n Accuracy : %.4f" % (model.evaluate(X, Y)[1])) 
```
