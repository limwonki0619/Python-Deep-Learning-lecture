## 필요 라이브러리 호출

```python
from sklearn.metrics import accuracy_score
from keras.datasets import cifar10
from keras.utils import to_categorical
from keras.models import Sequential, Model
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, Activation, Reshape, concatenate, Dropout
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.regularizers import l2

import numpy
import os
```

## 데이터 전처리

```python
# 데이터 로드 
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
# 총 데이터 

print(X_train.shape, X_train.dtype, X_train[0][0][0])
print(X_test.shape, y_train.dtype)
print(y_train.shape, X_test.dtype )
print(y_test.shape, y_test.dtype )
```

```python
X_train = X_train.astype('float32') / 255.0  # 0~1사이의 값으로 변환 
y_train = to_categorical(y_train)            # 원-핫 인코딩 
X_test = X_test.astype('float32') / 255.0
y_test = to_categorical(y_test)
```

```python
print(X_train.shape, X_train.dtype, X_train[0][0][0] )
```

## 데이터 확인

```python
# 이미지 그래프로 확인
import matplotlib.pyplot as plt

plt.subplot(141)
plt.imshow(X_train[6], interpolation="bicubic")
plt.subplot(142)
plt.imshow(X_train[60], interpolation="bicubic")
plt.subplot(143)
plt.imshow(X_train[600], interpolation="bicubic")
plt.subplot(144)
plt.imshow(X_train[6000], interpolation="bicubic")
plt.show()
```

## 모델 생성 및 컴파일

```python
model = Sequential([
    
    Conv2D(128, (5,5), padding='same', input_shape=(32, 32, 3), activation='relu', kernel_regularizer=l2(0.001)),
    Conv2D(128, (5,5), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.1),
    
    Conv2D(64, (5,5), padding='same', activation='relu', kernel_regularizer=l2(0.001)),
    Conv2D(64, (5,5), padding='same', activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.2),
    
    Conv2D(64, (5,5), padding='same', activation='relu', kernel_regularizer=l2(0.001)),
    Conv2D(64, (5,5), padding='same', activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.3),   
    
    Conv2D(64, (5,5), padding='same', activation='relu', kernel_regularizer=l2(0.001)),
    Conv2D(64, (5,5), padding='same', activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.4),   
    
    Flatten(),
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(.5), 
    Dense(10, activation='softmax')
])

model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='categorical_crossentropy', 
                        optimizer='adam', # 'adadelta'
                        metrics=['accuracy'])
```

## 모델 최적화

```python
MODEL_DIR = './CNN_cifar10_model/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)

modelpath="./CNN_cifar10_model/{epoch:02d}-{val_loss:.4f}.hdf5"
    
checkpoint = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10)
```

```python
history = model.fit(X_train, y_train, batch_size=100, epochs=100, validation_data=[X_test, y_test],
          verbose=2, callbacks=[early_stopping_callback, checkpoint])
```

## 모델 예측 정확도 

```python
# 테스트 정확도 출력
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, y_test)[1]))
```

## 모델 학습 과정 시각화

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# fm._rebuild() # <- 코드로 아나콘다에서 인식하지 못한 폰트를 업데이트 해준다.

font_fname = 'C:/Windows/Fonts/NanumGothic.ttf'  # 원하는 폰트 
font_family = fm.FontProperties(fname=font_fname).get_name()

%matplotlib inline 
```

```python
y_vloss = history.history['val_loss']   # 테스트 셋의 오차
y_loss = history.history['loss']        # 학습셋의 오차
x_len = numpy.arange(1, len(y_loss)+1)  # x축 
```

```python
# LOSS 그래프 생성 
plt.rcParams['font.family'] = font_family
plt.rcParams["font.size"] = 15
plt.figure(figsize=(12, 8))

plt.plot(x_len, y_vloss, marker='.', c="red", label='Testset_loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trainset_loss')

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch', size=18)
plt.ylabel('loss', size=18)
plt.show()
```
